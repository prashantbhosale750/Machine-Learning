# LDA Handson
    as LDA is supervised while transforming the data we need to consider Target variable

    üìå Steps for LDA Implementation
    1Ô∏è‚É£ Standardize the Data ‚Äì Ensure all features have the same scale.
    2Ô∏è‚É£ Compute the Mean Vectors ‚Äì Find the mean for each class.
    3Ô∏è‚É£ Compute the Within-Class Scatter Matrix (Sw) ‚Äì Measures spread within each class.
    4Ô∏è‚É£ Compute the Between-Class Scatter Matrix (Sb) ‚Äì Measures spread between classes.
    5Ô∏è‚É£ Compute Eigenvalues & Eigenvectors ‚Äì Find the best projection direction.
    6Ô∏è‚É£ Select Top k Eigenvectors ‚Äì Choose the components that capture the most class separability.
    7Ô∏è‚É£ Project Data onto New Axes ‚Äì Transform the original data into the new space.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
```

```python
#for the inbuild dataset
from sklearn.datasets import load_digits 
dataset=load_digits() 
```

```python
#splitting the data into indepedent and dependent
X=df
y=dataset["target"]
```

```python
#preprocessig of date : standardinze the data

from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
Scaled_X=sc.fit_transform(X)
```

```python
Scaled_X
```

```python
Scaled_X=pd.DataFrame(Scaled_X,columns=X.columns)
```

```python
Scaled_X
```

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
```

```python
lda_model=LinearDiscriminantAnalysis(n_components=9)
```

```python
lda_model.fit_transform(Scaled_X,y)
#passsing the target also along with independent data as lda uses these targets for seperability
```

```python
lda_X=pd.DataFrame(lda_X)
```

```python
lda_X #new indepennt column
```

```python
y #dependent
```

# model building 


```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test, y_train, y_test = train_test_split(lda_X, y, test_size=0.3,random_state=42)
```

```python
dt_model = DecisionTreeClassifier()
```

```python
#training of model
dt_model.fit(x_train, y_train) 
```

```python
# testing the model
y_pred = dt_model.predict(x_test) 
```

```python
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import *
```

```python
accuracy_score(y_test,y_pred)
```
